```python
def get_U(X):
    k = 2
    u, s, vh = np.linalg.svd(X)
    s[k-1:] = 0
    return u.dot(np.diag(s)).dot(vh)
```

```python
e = 1e-6
X = np.nan_to_num(A)
j = 0
while True:
    j += 1
    X_ = np.copy(X)
    X_k = get_U(X)
    for i in np.argwhere(np.isnan(A)):
        X[i] = X_k[i]
    print(j, np.linalg.norm(X - X_))
    if np.linalg.norm(X - X_) < e:
        break  
    
    
```
